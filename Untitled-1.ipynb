```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (8, 6)
sns.set(style='whitegrid')
np.random.seed(42)
```

```python
# Path to Titanic dataset
DATA_PATH = 'titanic.csv'

if os.path.exists(DATA_PATH):
    df = pd.read_csv(DATA_PATH)
else:
    # fallback: seaborn built-in titanic dataset
    df = sns.load_dataset('titanic')

# Clean column names
df.columns = [c.strip().lower() for c in df.columns]
df.head()
```

```python
print("Missing values per column:")
print(df.isna().sum())

plt.figure()
sns.barplot(x=df.isna().mean().values*100, y=df.isna().mean().index)
plt.title("Missing Values %")
plt.show()
```

```python
# Age distribution
plt.figure()
plt.hist(df['age'].dropna(), bins=30, edgecolor='k')
plt.title('Age Histogram')
plt.show()

# Fare distribution with KDE
plt.figure()
sns.histplot(df['fare'].dropna(), bins=40, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Simulate a normal distribution with randn
sim = np.random.randn(1000)
plt.hist(sim, bins=30, edgecolor='k')
plt.title('Simulation: np.random.randn')
plt.show()
```

```python
# Drop columns with more than 40% missing values
missing_ratio = df.isna().mean()
df.drop(columns=missing_ratio[missing_ratio > 0.4].index, inplace=True)

# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Fill embarked with mode
if 'embarked' in df.columns:
    df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Fill age with median (grouped by pclass & sex if available)
if 'age' in df.columns:
    if 'pclass' in df.columns and 'sex' in df.columns:
        df['age'] = df['age'].fillna(df.groupby(['pclass','sex'])['age'].transform('median'))
    df['age'].fillna(df['age'].median(), inplace=True)

# Fill fare with median (grouped by pclass if available)
if 'fare' in df.columns:
    if 'pclass' in df.columns:
        df['fare'] = df['fare'].fillna(df.groupby('pclass')['fare'].transform('median'))
    df['fare'].fillna(df['fare'].median(), inplace=True)

df.head()
```

```python
def iqr_outlier_indexes(series):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    return series[(series < lower) | (series > upper)].index

# Remove outliers from Age and Fare
for col in ['age','fare']:
    if col in df.columns:
        df.drop(index=iqr_outlier_indexes(df[col].dropna()), inplace=True)

df.reset_index(drop=True, inplace=True)
df.shape
```

```python
# Family size = siblings/spouses + parents/children + self
if 'sibsp' in df.columns and 'parch' in df.columns:
    df['familysize'] = df['sibsp'] + df['parch'] + 1

# Encode categorical variables
if 'sex' in df.columns:
    df['sex_num'] = df['sex'].map({'male':1,'female':0})

if 'embarked' in df.columns:
    df['embarked_num'] = df['embarked'].astype('category').cat.codes

df.head()
```

```python
# Fare vs Pclass
if set(['pclass','fare']).issubset(df.columns):
    sns.boxplot(x='pclass', y='fare', data=df)
    plt.title('Fare vs Pclass')
    plt.show()

# Survival by Sex
if set(['sex','survived']).issubset(df.columns):
    sns.countplot(x='sex', hue='survived', data=df)
    plt.title('Survival by Sex')
    plt.show()

# Survival by Pclass
if set(['pclass','survived']).issubset(df.columns):
    sns.countplot(x='pclass', hue='survived', data=df)
    plt.title('Survival by Pclass')
    plt.show()

# Age distribution by Survival
if set(['age','survived']).issubset(df.columns):
    sns.kdeplot(x='age', hue='survived', data=df, fill=True, alpha=0.4)
    plt.title('Age Distribution by Survival')
    plt.show()

# Age vs Fare scatter by Survival
if set(['age','fare','survived']).issubset(df.columns):
    sns.scatterplot(x='age', y='fare', hue='survived', data=df)
    plt.title('Age vs Fare by Survival')
    plt.show()
```

```python
# Correlation heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# Final pairplot
num_cols = ['age','fare','familysize']
num_cols = [c for c in num_cols if c in df.columns]

if num_cols:
    sns.pairplot(df[num_cols + ['survived']], hue='survived', diag_kind='kde')
    plt.show()
```
